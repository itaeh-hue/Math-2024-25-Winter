1. Find the equation of a circle with center at point $A(1,2)$ and radius $r=3$.


The equation of a circle with center at $(h, k)$ and radius $r$ is given by:

$$(x - h)^2 + (y - k)^2 = r^2$$

For our circle at $A(1,2)$:

$$h=1 \quad k=2 \quad r =3$$$$(x - 1)^2 + (y - 2)^2 = 3^2 \implies$$
$$(x - 1)^2 + (y - 2)^2 = 9$$

[GeoGebra_link](https://www.geogebra.org/m/uz85nux5)

```python
import numpy as np
import matplotlib.pyplot as plt

# Circle parameters
h, k = 1, 2  # center
r = 3         # radius

# Create points for circle
theta = np.linspace(0, 2 * np.pi, 100)
x = h + r * np.cos(theta)
y = k + r * np.sin(theta)

# Plotting
plt.figure(figsize=(6,6))
plt.plot(x, y, label='Circle: $(x - 1)^2 + (y - 2)^2 = 9$')
plt.scatter(h, k, color='red')  # center
plt.text(h, k, ' A(1,2)', fontsize=12, verticalalignment='bottom')
plt.xlim(-5, 8)
plt.ylim(-5, 8)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Circle with Center A(1, 2) and Radius 3")
plt.grid()
plt.legend()
plt.show()
```

2. Find the equation of a parabola intersecting the $Ox$ axis at points $x=2$, $x=4$, and passing through point $y(3)=1$.


The general form of a parabola that intersects the $x$-axis at points $a$ and $b$ can be written as:

$$y = k(x - a)(x - b)$$

- $a = 2$
- $b = 4$

Therefore,

$$y = k(x - 2)(x - 4)$$

Given the fact that the parabola passes through the point $(3, 1)$:

$$1 = k(3 - 2)(3 - 4) \implies 1 = k(1)(-1) \implies 1 = -k \implies k = -1$$

Substituting $k$ back into the equation:

$$y = -1(x - 2)(x - 4) = -(x^2 - 6x + 8) = -x^2 + 6x - 8$$

Final equation:

$$y = -x^2 + 6x - 8$$

3. Find the center of the ellipse with the equation $x^2 + 4y^2 - 4x - 16y + 16 = 0$.

First, rearrange the equation and group the terms:
$$(x^2 - 4x) + (4y^2 - 16y) + 16 = 0$$

Then complete the square for $x$:
$$(x^2 -4x) = ((x - 2)^2 - 4)$$
And for $y$:
$$4(y^2 - 4y) = 4(y^2 - 4y + 4 - 4) = 4((y - 2)^2 - 4) = 4(y - 2)^2 - 16$$

This implies

$$(x^2 - 4x) + (4y^2 - 16y) + 16 = 0 \implies$$
$$((x - 2)^2 - 4) + 4(y - 2)^2 - 16 + 16 = 0$$
$$(x - 2)^2 + 4(y - 2)^2 -4 = 0$$

Adding $4$ to both sides:
$$(x - 2)^2 + 4(y - 2)^2 = 4$$

Dividing by $4$:

$$\frac{(x - 2)^2}{4} + \frac{(y - 2)^2}{1} = 1$$
The center is given by $(2,2)$.

4. Find the slope ($m>0$) of the line $y=mx-5$ that is tangent to the circle with the equation $x^2 + y^2=1$.

To find the slope of the tangent line, we need to ensure that the line does not intersect the circle at more than one point.

Substituting $y = mx - 5$ into the circle's equation:

$$x^2 + (mx - 5)^2 = 1$$

Expanding:

$$x^2 + (m^2 x^2 - 10mx + 25) = 1$$

Combining terms:

$$(1 + m^2)x^2 - 10mx + 24 = 0$$

For the line to be tangent to the circle, the discriminant of this quadratic must be zero:

$$D = b^2 - 4ac = (-10m)^2 - 4(1 + m^2)(24) = 0$$

Calculating the discriminant:

$$100m^2 - 96 - 96m^2 = 0 \Rightarrow 4m^2 - 96 = 0$$

Solving for $m^2$:

$$4m^2 = 96 \implies m^2 = 24 \implies m = 2\sqrt{6}$$

Since we want $m > 0$:

Final Slope: $m = 2\sqrt{6}$.

5. Find the intersection points of the hyperbola $x^2 - y^2 = 1$ with the ellipse's line $x^2 + 4y^2 = 6$.

We substitute
$$y^2 = x^2 - 1$$
into the eclipse's equation:

$$x^2 + 4(x^2 - 1) = 6$$

$$x^2 + 4x^2 - 4 = 6 \implies 5x^2 - 4 = 6 \implies 5x^2 = 10 \implies x^2 = 2 \implies x = \pm \sqrt{2}$$

Now substituting $x = \sqrt{2}$ into the hyperbola's equation to find $y$:

$$y^2 = (\pm \sqrt{2})^2 - 1 = 2 - 1 = 1 \implies y = \pm 1$$


So we have four points:

1. $(\sqrt{2}, 1)$
2. $(\sqrt{2}, -1)$
3. $(-\sqrt{2}, -)$
4. $(-\sqrt{2}, -1)$
