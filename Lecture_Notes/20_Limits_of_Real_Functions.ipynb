## Introduction to Limits

The concept of a limit is fundamental in calculus and analysis. A limit describes the behavior of a function as its input approaches a particular point from either side. Limits are crucial for understanding continuity, derivatives, and integrals.

### Definition

For a function $f(x)$, the limit as $x $approaches $a $is denoted as:

$$
\lim_{x \to a} f(x) = L
$$

This notation means that as $x $gets arbitrarily close to $a$, the values of $f(x) $get arbitrarily close to $L$.

## Techniques for Evaluating Limits

### a. Direct Substitution

Many limits can be computed by direct substitution of the point into the function. If $f(a) $exists, then:

$$
\lim_{x \to a} f(x) = f(a)
$$

### b. Factoring and Simplifying

If direct substitution results in an indeterminate form (like $\frac{0}{0}$), we can often resolve this by factoring and simplifying the expression.

### c. L'Hôpital's Rule

For indeterminate forms such as $\frac{0}{0} $or $\frac{\infty}{\infty}$, we can apply L'Hôpital's Rule, which states:

$$
\lim_{x \to a} \frac{f(x)}{g(x)} = \lim_{x \to a} \frac{f'(x)}{g'(x)}
$$

provided that the limit on the right exists.

### d. Squeeze Theorem

If a function $f(x) $is squeezed between two other functions $g(x) $and $h(x) $(i.e., $g(x) \leq f(x) \leq h(x)$) and the limits of $g(x) $and $h(x) $are equal as $x $approaches $a$:

$$
\lim_{x \to a} g(x) = \lim_{x \to a} h(x) = L
$$

Then:

$$
\lim_{x \to a} f(x) = L
$$

## Asymptotes

Asymptotes are lines that a graph approaches as $x $or $y $approaches infinity. They help us understand the end behavior of functions.

### a. Vertical Asymptotes

Vertical asymptotes occur where a function approaches infinity or negative infinity. They typically exist at values of $x $where the function is undefined and the limit approaches $\pm \infty$.

To identify vertical asymptotes, set the denominator of a rational function to zero and check whether the limits:

$$
\lim_{x \to a^-} f(x) \quad \text{and} \quad \lim_{x \to a^+} f(x)
$$

both approach $\pm \infty$.

### b. Horizontal Asymptotes

Horizontal asymptotes provide information about the behavior of a function as $x $approaches infinity or negative infinity.

To find horizontal asymptotes, evaluate:

$$
\lim_{x \to \infty} f(x) \quad \text{and} \quad \lim_{x \to -\infty} f(x)
$$

The results indicate whether $f(x) $approaches a finite value $L $or diverges.

- If the limit equals a finite number as $x \to \infty$, there is a horizontal asymptote at $y = L$.
- If the limit equals $0$, then there is a horizontal asymptote at $y = 0$.

### Examples of Finding Asymptotes

1. **Vertical Asymptotes**: For the function $f(x) = \frac{1}{x - 2}$, a vertical asymptote occurs at $x = 2 $since the function is undefined there.
   
2. **Horizontal Asymptotes**: For $f(x) = \frac{3x^2 + 5}{2x^2 + 1}$, we find:

$$
\lim_{x \to \infty} f(x) = \frac{3}{2}
$$

This indicates a horizontal asymptote at $y = \frac{3}{2}$.

## Summary of Key Concepts

- **Limit**: Determines the behavior of functions as inputs approach specific points.
- **Vertical Asymptotes**: Indicate points of discontinuity where function outputs diverge to infinity.
- **Horizontal Asymptotes**: Describe the end behavior of functions as inputs extend towards infinity.

Understanding these concepts is essential for analyzing functions, particularly in calculus and higher mathematics.

   - $\displaystyle\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}$

Solution:
$$\lim_{ x \to \infty } \frac{x^3 + 2x^2}{x^4 - 3x^3} = \lim_{ x \to \infty } \frac{x^4\left( \frac{1}{x} + \frac{2}{x^2} \right)}{x^4\left( 1 - \frac{3}{x} \right)} = \frac{0}{1} = 0$$

Solution:

$$\lim_{ x \to 0 } \frac{\sin 3x}{2x+1} = \lim_{ x \to 0 } \frac{\sin 3x}{2x+1} \cdot \frac{3x}{3x} = \lim_{ x \to 0 } \frac{\sin 3x}{3x} \cdot \frac{3x}{2x+1} = 1$$

### Exercise 3

>3. Find the asymptotes of the function:
  
   - $f(x) = \frac{x^2 - 1}{x^2 + 1}$
   - $g(x) = \frac{\sin(x)}{x^2+1}$


   - For $f(x) = \frac{x^2 - 1}{x^2 + 1}$:

Vertical asymptotes occur where the function is undefined, typically where the denominator is zero. We check the denominator: $x^2 + 1 = 0$ . However, this equation has no real solutions, since the square can not be negative. Therefore, there are no vertical asymptotes for $f(x)$.

To find horizontal asymptotes, we examine the behavior of $f(x)$ as $x$ approaches infinity:

$$\lim_{ x \to \infty } f(x) = \frac{x^2 - 1}{x^2 + 1} = \lim_{ x \to \infty } \frac{1 - \frac{1}{x^2}}{1 + \frac{1}{x^2}} =1$$
Thus, there is a horizontal asymptote at $y = 1$.

 - For $g(x) = \frac{\sin x}{x^2 + 1}$:

Vertical Asymptotes:  $g(x) = \frac{\sin x}{x^2 + 1}$
Again, we check where the function $g(x)$ is undefined, which would occur where the denominator is zero: $x^2 + 1 = 0$. Like before: no real solutions - no vertical asymptotes.

To find horizontal asymptotes, we evaluate the limit of $g(x)$ as $x \to \infty$ (and also $x \to -\infty$):

$$
\lim_{ x \to \infty } g(x) = \frac{\sin x}{x^2 + 1}
$$

Since $|\sin x| \leq 1$:

$$\left| \frac{\sin x}{x^2 + 1} \right| \leq \frac{1}{x^2 + 1}
$$

As $x \to \infty$, $\frac{1}{x^2 + 1} \to 0$, so by the Squeeze Theorem:

$$\lim_{x \to \infty} g(x) = 0$$

Therefore, there is a horizontal asymptote at $y = 0$.
