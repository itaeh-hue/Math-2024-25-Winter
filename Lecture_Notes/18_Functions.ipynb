## Functions

>**Definition. Functions.**

>Let $X, Y$ be sets, and $P(x,y)$: $$\forall x\in X, \exists y \in Y: P(x,y) \text{ is true, and there is only one such $y$.}$$

>We define the *function* $f:X \rightarrow Y$ *defined by $P$ on the domain $X$ and range $Y$* to be the object which, given any input $x\in X$, assigns an output $f(x)\in Y$, defined to be the unique object $f(x)$ for which $P(x, f(x))$ is true.
>Thus, $\forall x\in X, y\in Y$, $$y = f(x) \iff P(x,y) \text{ is true.}$$


>**Definition. Equality of functions.**

>Two functions $f:X \rightarrow Y, g:X\rightarrow Y$ with the same domain and range are said to be *equal*, $f=g$, if and only if $f(x) = g(x), \forall x\in X$.

### Function compositions

>**Definition. Composition.**

>Let $f:X \rightarrow Y, g:Y\rightarrow Z$ be two functions, such that the range of $f$ is the same set as the domain of $g$. We then define the *composition* $g \circ f:X\rightarrow Z$ of the two functions $g$ and $f$ to be the function defined explicitly by the formula $$(g\circ f)(x) := g(f(x))$$
>If the range of $f$ does not match the domain of $g$, we leave the composition $g \circ f$ undefined.


>**Lemma. Composition is associative.**

>Let $h : X → Y$, $g : Y → Z$, and $f : Z → W$ be functions. Then $$f \circ (g \circ h) = (f \circ g) \circ h.$$


### Injective, srujective, and bijective functoins

>**Definition. One-to-one functions.**

>A function $f$ is *one-to-one*, aka *injective*, if $$x\neq x'\implies f(x)\neq f(x')$$
>equivalently, a function is one-to-one if $$f(x)= f(x') \implies x=x'$$


---

>**Definition. Onto functions.**

>A function $f$ is *onto*, or *surjective*, if $$\forall y\in Y, \exists x\in X: f(x) = y$$
>In other words, every element $y$ in $Y$ comes from applying $f$ to some element in $X$.

---

>**Definition. Bijective functions.**

>Functions $f : X → Y$ which are both one-to-one and onto are also called *bijective*, or *invertible*.

#### Functions as ordered sets

Functions can also be defined as sets of ordered pairs, $f = \{(x_1, y_1), (x_2, y_2), ... \}$.

- If each output $y$ corresponds to a unique input $x$, then an inverse can be found by switching the order of the pairs. For example, if $f = \{(1, 7), (2, 9), (3, 11)\}$, then the inverse function $f^{-1} = \{(7, 1), (9, 2), (11, 3)\}$.
