## Exercises

Solve the following systems of equations without using matrices:

1. $3x-2y=5, \quad 2x+3y=7$,
2. $2x-3y=10, \quad 4x+5y=20$,
3. $2x - y + z = 3, \quad x + 2y - z = 1, \quad 3x - y + 2z = 11$.
4. $2x-3y+4z+2t=2$, $3x+2y-5z+3t=3$, $4x-3y+2z-5t=4$, $5x+4y-3z+2t=5$.




### Exercise 1

Given $$\begin{cases}
3x - 2y &= 5 \quad \\
2x + 3y &= 7 \quad
\end{cases}$$
From the first equation, $3x - 2y = 5 \implies 2y = 3x - 5 \implies y = \frac{3x - 5}{2}$; then substitute this value into the second equation: $2x + 3\left(\frac{3x - 5}{2}\right) = 7 \implies 4x + 3(3x - 5) = 14 \implies 4x + 9x - 15 = 14 \implies 13x - 15 = 14 \implies 13x = 29 \implies x = \frac{29}{13}$

This value is then again substituted to find $y$: $y = \frac{3x - 5}{2} =  y = \frac{3\cdot \frac{29}{13} - 5}{2} =  \frac{\frac{87}{13} - \frac{65}{13} }{2} = \frac{\frac{22}{13} }{2} = \frac{11}{13}$.

So that, final solution: $$(x, y) = \left(\frac{29}{13}, \frac{11}{13}\right)$$
### Exercise 2

Given $$\begin{cases}
2x - 3y &= 10 \quad\\
4x + 5y &= 20 \quad
\end{cases}$$

From the first one, $2x = 3y + 10 \implies x = \frac{3y + 10}{2}$; then $4\left(\frac{3y + 10}{2}\right) + 5y = 20\implies 22y + 40 = 40 \implies 22y = 0 \implies y = 0$. Then $x = \frac{3y + 10}{2} = 5$.

Final solution: $$(x, y) = (5, 0)$$
### Exercise 3

Given $$
\begin{cases}
2x - y + z &= 3 \quad \text{(1)}\\
x + 2y - z &= 1 \quad \text{(2)}\\
3x - y + 2z &= 11 \quad \text{(3)}
\end{cases}$$
1. Solve $(1)$ for $z$: ${z = 3 - 2x + y}$
2. Substitute $z$ into $(2)$: $x + 2y - (3 - 2x + y) = 1 \implies 3x + y = 4$
3. Substitute $z$ into $(3)$: $3x - y + 2(3 - 2x + y) = 11 \implies -x + y = 5 \implies {y = x+5}$
4. Substitute $y$ to find $x$ from step 2.: $3x + y = 4\implies 3x + x+5 = 4\implies 4x = -1\implies \boxed{x=-\frac{1}{4}}$
5. Then this $x$ goes back to $z$ from step 1.: $z = 3-2x + y \implies z=3-2\left(  -\frac{1}{4} \right) + \left(  -\frac{1}{4} +5 \right)$; $z=3+ \frac{2}{4} - \frac{1}{4} +5=8 \frac{1}{4}= \frac{33}{4}$; $\boxed{z= \frac{33}{4}}$
6. $y = x + 5 = -\frac{1}{4} + 5 = \frac{19}{4}$; $\boxed{y = \frac{19}{4}}$

Final solution: $$\left(x, y, z\right) = \left(-\frac{1}{4}, \frac{19}{4}, \frac{33}{4}\right)$$
### Exercise 4

Given

$$\begin{cases}
2x - 3y + 4z + 2t &= 2 \quad \text{(1)}\\
3x + 2y - 5z + 3t &= 3 \quad \text{(2)}\\
4x - 3y + 2z - 5t &= 4 \quad \text{(3)}\\
5x + 4y - 3z + 2t &= 5 \quad \text{(4)}
\end{cases}$$


Step 1: Solve Equation (1) for $t$
From (1):
$$
2t = 2 - 2x + 3y - 4z \implies t = 1 - x + \frac{3}{2}y - 2z \quad \text{(5)}
$$

Step 2: Substitute (5) into Equations (2), (3), and (4)
1. Substituting (5) into (2):
   $$
   3x + 2y - 5z + 3\left(1 - x + \frac{3}{2}y - 2z\right) = 3
   $$
   Simplify:
   $$
   3x + 2y - 5z + 3 - 3x + \frac{9}{2}y - 6z = 3 \implies \frac{13}{2}y - 11z = 0 \quad \text{(6)}
   $$

2. Substituting (5) into (3):
   $$
   4x - 3y + 2z - 5\left(1 - x + \frac{3}{2}y - 2z\right) = 4
   $$
   Simplify:
   $$
   4x - 3y + 2z - 5 + 5x - \frac{15}{2}y + 10z = 4 \implies 9x - \frac{21}{2}y + 12z = 9 \quad \text{(7)}
   $$

3. Substituting (5) into (4):
   $$
   5x + 4y - 3z + 2\left(1 - x + \frac{3}{2}y - 2z\right) = 5
   $$
   Simplify:
   $$
   5x + 4y - 3z + 2 - 2x + 3y - 4z = 5 \implies 3x + 7y - 7z = 3 \quad \text{(8)}
   $$

Step 3: Solve the new system of equations (6), (7), (8)
From (6):
$$
\frac{13}{2}y = 11z \implies y = \frac{22}{13}z \quad \text{(9)}
$$

Substituting (9) into (7):
$$
9x - \frac{21}{2}\left(\frac{22}{13}z\right) + 12z = 9
$$
Multiply through by 26 to eliminate fractions:
$$
234x - 231z + 312z = 234 \implies 234x + 81z = 234
$$
Thus:
$$
x + \frac{27}{78}z = 1 \implies x = 1 - \frac{27}{78}z \quad \text{(10)}
$$

Substituting (9) and (10) into (8):
$$
3\left(1 - \frac{27}{78}z\right) + 7\left(\frac{22}{13}z\right) - 7z = 3
$$
Multiply through by 78:
$$
234 - 81z + 118z - 546z = 234
$$
Combine:
$$
234 - 509z = 234 \implies -509z = 0 \implies z = 0
$$

Step 4: Find $y$ and $x$
Using (9) with $z = 0$:
$$
y = \frac{22}{13}(0) = 0
$$
Using (10) with $z = 0$:
$$
x = 1 - \frac{27}{78}(0) = 1
$$
Finally, using (5):
$$
t = 1 - 1 + \frac{3}{2}(0) - 2(0) = 1
$$

Final Solution:
$$
\boxed{(x, y, z, t) = (1, 0, 0, 1)}
$$
