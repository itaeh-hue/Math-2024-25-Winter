## Integrals

Integrals are fundamental mathematical tools used in calculus to measure areas, volumes, and quantities that accumulate over time or space. The process of integration is essentially the reverse of differentiation. There are two main types of integrals:

1. **Indefinite Integrals**: used to calculate anti-derivatives; does not have specified limits of integration.
2. **Definite Integrals**: used to calculate the net area under a curve between two specific points; does have specified limits of integration.

### **Indefinite Integrals**

The notation for an indefinite integral is given by:

$$
\int f(x) \, dx
$$

where $f(x)$ is a function of $x$. The result of this integral is a family of functions $F(x)$ such that $F'(x) = f(x)$, along with the constant of integration $C$:

$$
\int f(x) \, dx = F(x) + C
$$

### Basic Properties of Integrals

1. **Linearity**: The integral operator is linear, meaning:
   $$
   \int (a f(x) + b g(x)) \, dx = a \int f(x) \, dx + b \int g(x) \, dx
   $$
   where $a$ and $b$ are constants.

2. **Power Rule**: For any real number $n \neq -1$:
   $$
   \int x^n \, dx = \frac{x^{n+1}}{n+1} + C
   $$

3. **Basic Functions**:
   - $\int 1 \, dx = x + C$
   - $\int e^{kx} \, dx = \frac{1}{k} e^{kx} + C$
   - $\int \sin(kx) \, dx = -\frac{1}{k} \cos(kx) + C$
   - $\int \cos(kx) \, dx = \frac{1}{k} \sin(kx) + C$
   - $\int \frac{1}{x} \, dx = \ln |x| + C$

4. **Trigonometric Identities**: Knowing these identities can help simplify integrals involving trigonometric functions:
   - $\sin^2(x) + \cos^2(x) = 1$

### Techniques of Integration

1. **Substitution Method**: Useful for more complex integrals, it involves changing the variable to simplify the integral:
   Let $u = g(x)$, then:
   $$
   \int f(g(x)) g'(x) \, dx = \int f(u) \, du
   $$

2. **Integration by Parts**: Based on the product rule for differentiation, this technique is useful for integrating products of functions:
   $$
   \int u \, dv = uv - \int v \, du
   $$

3. **Definite Integrals**: Given by:
   $$
   \int_a^b f(x) \, dx = F(b) - F(a)
   $$
   where $F(x)$ is any antiderivative of $f(x)$.

### Visualization of Integrals

Calculating definite integrals often involves visually interpreting the area under a curve. Tools such as GeoGebra or graphing calculators can help illustrate these areas and the functions involved.

### Applications of Integrals

1. **Area Under Curves**: The area between a function and the x-axis can be computed using definite integrals.
2. **Physical Applications**: In physics, integrals are used to find quantities like distance from velocity functions, work done from the force applied, and the center of mass.

## Examples of Common Integrals

2. $\int x^n \, dx = \frac{x^{n+1}}{n+1} + C$ for $n \neq -1$
3. $\int e^{kx} \, dx = \frac{1}{k} e^{kx} + C$
4. $\int \sin(kx) \, dx = -\frac{1}{k} \cos(kx) + C$
