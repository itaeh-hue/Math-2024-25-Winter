## Derivatives

The derivative of a function provides a measure of how the function value changes as its input changes. It is defined formally as:

$$
f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
$$

This definition can be interpreted as the slope of the tangent line to the curve at the point $x$.

### Basic Derivative Rules

- **Constant Rule**: If $c$ is a constant, then:
  $$
  \frac{d}{dx}(c) = 0
  $$

- **Power Rule**: If $n$ is a real number, then:
  $$
  \frac{d}{dx}(x^n) = n x^{n-1}
  $$

- **Constant Multiple Rule**: If $c$ is a constant and $f(x)$ is a function, then:
  $$
  \frac{d}{dx}[c f(x)] = c f'(x)
  $$

- **Sum Rule**: For functions $f(x)$ and $g(x)$:
  $$
  \frac{d}{dx}[f(x) + g(x)] = f'(x) + g'(x)
  $$

- **Difference Rule**:
  $$
  \frac{d}{dx}[f(x) - g(x)] = f'(x) - g'(x)
  $$


- **Product Rule**: If $f(x)$ and $g(x)$ are functions, then:
  $$
  \frac{d}{dx}[f(x)g(x)] = f'(x)g(x) + f(x)g'(x)
  $$

- **Quotient Rule**: If $f(x)$ and $g(x)$ are functions and $g(x) \neq 0$, then:
  $$
  \frac{d}{dx}\left[\frac{f(x)}{g(x)}\right] = \frac{f'(x)g(x) - f(x)g'(x)}{[g(x)]^2}
  $$

- **Higher-Order Derivatives**: The second derivative is denoted $f''(x)$ and is found by differentiating $f'(x)$ again.
- **Implicit Differentiation**: Used when you have an equation in $x$ and $y$ that cannot be easily solved for $y$.

The derivatives of the basic trigonometric functions are as follows:

- **Trigonometric functions**:
	- $\frac{d}{dx}\sin x = \cos x$
	- $\frac{d}{dx}\cos x = -\sin x$
	- $\frac{d}{dx}(\tan(x)) = \sec^2(x)$

- **Exponential Functions**:
	- $\frac{d}{dx}(e^x) = e^x$
	- General case: $\frac{d}{dx}(a^x) = a^x \ln(a)$, where $a > 0$.
  
- **Logarithmic Functions**:
	- $\frac{d}{dx}(\ln(x)) = \frac{1}{x}$ for $x > 0$
	- For $\ln(g(x))$:
  $$
  \frac{d}{dx}(\ln(g(x))) = \frac{g'(x)}{g(x)}
  $$

### Chain Rule

The chain rule is used to differentiate composite functions. If $y = g(u)$ and $u = f(x)$, then:
$$
\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx} = g'(f(x)) \cdot f'(x)
$$





>The second derivative $f''(x)$ is the derivative of the derivative. It provides information about the curvature of the function and is used in determining concavity.
### L'Hospital's Rule

If you have the limit $\lim_{x \to c} \frac{f(x)}{g(x)}$ that results in an indeterminate form $\frac{0}{0}$ or $\frac{\infty}{\infty}$, then:

$$
\lim_{x \to c} \frac{f(x)}{g(x)} = \lim_{x \to c} \frac{f'(x)}{g'(x)} \quad \text{(if this limit exists)}
$$

This allows us to differentiate the numerator and denominator until the limit can be evaluated.

#### 4. $y(x) = 2x^3 - 3x^2 + 8x - 9$

$y'(x) = (6 x^{1/3})'= 2x^{-2/3} = \frac{2}{\sqrt[3]{x^2}}$

$y'(x) = (\cos x + \sin x)'=-\sin x + \cos x$

$y'(x) = (2\sin x \cos x)'=2(\sin x\cos x)'=2(\sin'x\cos x+ \cos 'x \sin x)=2(\cos x \cos x - \sin x \sin x) = 2\cos^2(x) - 2\sin^2(x) = \sin(2x)$
#### 9. $y(x) = x\sin x$

$\frac{d}{dx}[f(x)g(x)] = f'(x)g(x) + f(x)g'(x)$

$y'(x) = (x\sin x)'=x'\sin x + x\sin'x=\sin x+x\cos x$
#### 10. $y(x) = (x+1)(x+1)$

$\frac{d}{dx}[f(x)g(x)] = f'(x)g(x) + f(x)g'(x)$

$y'(x) = ((x+1)(x+1))'=(x+1)'(x+1)+(x+1)'(x+1)=(x+1)+(x+1)=2x+2$

Using the chain rule,  $(g(f(x)))' = g'(f(x)) f'(x)$, we can easily find:

### Exercise 5

>5. In physics, the position of a particle is given by $x(t) = 3t^2 - 6t + 1$. Find the velocity $V(t)=x'(t)$ and acceleration $a(t)=V'(t)=x''(t)$ of the particle at time $t = 2$.

Velocity, being the first derivative, can be computed like this:

$$x'(t) = (3t^2 - 6t + 1)'=6t-6$$

And acceleration is then:

$$x''(t) - (6t-6)'=6$$
