## Differential equations

>A differential equation is an equation that relates one or more unknown functions and their derivatives.

$$y' = f(x)$$

Differential equations are categorized into several types based on their properties and characteristics:

- **Ordinary Differential Equations (ODEs)**
	- These equations involve derivatives with respect to a single independent variable. They are further classified into linear and nonlinear ODEs.

- **Partial Differential Equations (PDEs)**
	- These equations involve partial derivatives with respect to multiple independent variables. They are also divided into linear and nonlinear PDEs.

>The order of a differential equation is defined as the highest order of the derivative of the dependent variable with respect to the independent variable present in the equation.

For example, if a differential equation includes the **second** derivative of the dependent variable, it is classified as a **second-order** differential equation.
## First-Order Ordinary Differential Equations (FOODEs)

>A First-Order Ordinary Differential Equation (FOODE) is an ODE in which the highest order derivative of the unknown function is first. Mathematically, a FOODE can be expressed as $$F(x, y, y') = 0$$ or $$y' = F(x, y)$$

We can classify FOODEs into two main categories:

- Separable ODEs
	- $y' = F(x, y)$, which can be separated into two functions, one of which depends on $x$ and the other on $y$.

- Non-separable ODEs
	- $F(x, y, y') = 0$, which cannot be separated into two functions.
### Solving Separable ODEs: Separation of Variables

The Separation of Variables method is used to solve FOODEs in which the unknown function can be separated into two functions. We rewrite the $y' = F(x, y)$ ODE as:

$$\frac{dy}{dx} = F(x, y)$$

and multiply both sides by the integrating factor, $M(x)$:

$$M(x) \frac{dy}{dx} = M(x) F(x, y)$$

The integrating factor is defined as:

$$M(x) = e^{\int \frac{1}{F(x, y)} dx}$$

We then integrate both sides to obtain the general solution.

#### $y'(x)= y$

This is a first-order linear ordinary differential equation that can be solved using **separation of variables**.

1. Separate variables:

$$\frac{dy}{dx}=y \implies \frac{dy}{y}=dx$$

Let us stop a bit. In $\frac{dy}{dx} =ky$, $\frac{dy}{dx}$ is a rate of change of $y$ proportional to $y$. In our case, $k=1$.

$\frac{dy}{dx} =ky \implies \frac{1}{y} \frac{dy}{dx}=k$. Now, if we integrate both sides with respect to $x$, we get $\int \frac{1}{y} \frac{dy}{dx} dx=\int kdx$. So, $dx$ on the left cancels and we get $\int \frac{1}{y} dy=\int kdx$. Evaluating the integral on both sides, we get: $\ln|y| =kx+C$, where $C$ is a constant (integral is without boundaries).

To get rid of the logarithm, we use $e$: $e^{\ln|y|} =e^{kx +C}$, and then $|y| = e^{kx+C}$, then $|y| = e^{kx}e^C$. Now, if $C$ is a constant, then $e^C$ is a constant as well, we can mark it $C_0$ and get $|y| = C_{0}e^{kx}$.

This is simplified, but anyway.

In our case:

$$y(x)=C_{0}e^x$$

#### $y'(x) = \frac{1}{2y(x)}$

 $y'(x) = \frac{1}{2y(x)}$ can be rewritten as $\frac{dy}{dx} =\frac{1}{2y} | \cdot 2y \implies 2y \frac{dy}{dx} =1 |\cdot dx \implies$$2y dy =dx$

Integrating both sides:

$$\int 2y dy =\int dx \implies y^2 = x+C$$
Solving for $y$:

$$|y| = \sqrt{ x + C }$$

#### $\frac{dy}{dx} = \frac{x}{y}$

$\frac{dy}{dx} = \frac{x}{y}$ $\implies$ $y \frac{dy}{dx} = x$, integrating both sides:

$$\int y \frac{dy}{dx} = \int x \implies \frac{y^2}{2} = \frac{x^2}{2} + C \implies y^2 = x^2 + 2C$$

So the general solution would be

$$|y| = \sqrt{ x^2 + 2C }$$

#### $\frac{dy}{dx} = \frac{y}{x}$

$\frac{dy}{dx} = \frac{y}{x} \implies dy = \frac{ydx}{x}\implies \frac{dy}{y}=\frac{dx}{x}$, integrating:

$$\int\frac{dy}{y}=\int\frac{dx}{x} \implies \ln|y|=\ln|x| + C \implies |y| = e^{C}|x| \implies y=Cx$$

#### $\frac{dy}{dx} = xy$

$\frac{dy}{dx} = xy$ $\implies dy = xydx \implies \frac{dy}{y} =xdx$, integrating:

$$\int \frac{1}{y}dy = \int xdx \implies \ln|y| = \frac{x^2}{2} + C \implies |y| = Ce^{\frac{x^2}{2}}$$
