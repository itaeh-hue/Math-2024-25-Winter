### Exercise 1

Given: $$\begin{cases}
x + 2y - 2z = 4 \\
2x + y + z = 0 \\
3x + 2y + z = 1
\end{cases}$$

To solve a system of linear equations using Gauss elimination method, we first turn the system into an augmented matrix, and then row-reduce it to a Row Echelon Form (REF):

$$\begin{bmatrix}
1 & 2 & -2 & | & 4 \\
2 & 1 &  1 & | & 0 \\
3 & 2 &  1 & | & 1
\end{bmatrix}R_{2}-2R_{1}\implies\begin{bmatrix}
1 & 2 & -2 & | & 4 \\
0 & -3 &  5 & | & -8 \\
3 & 2 &  1 & | & 1
\end{bmatrix}R_{3}-3R_{1}\implies\begin{bmatrix}
1 & 2 & -2 & | & 4 \\
0 & -3 &  5 & | & -8 \\
0 & -4 &  7 & | & -11
\end{bmatrix}R_3 - \frac{4}{3}R_2\implies\begin{bmatrix}
1 & 2 & -2 & | & 4 \\
0 & -3 &  5 & | & -8 \\
0 & 0 &  \frac{1}{3} & | & -\frac{1}{3}
\end{bmatrix}R_{3}\cdot 3\implies \begin{bmatrix}
1 & 2 & -2 & | & 4 \\
0 & -3 &  5 & | & -8 \\
0 & 0 &  1 & | & -1
\end{bmatrix}$$

From the results, we have:

1. $z = \boxed{-1}$
2. $-3y + 5z = -8\implies -3y -5=-8\implies -3y=-3\implies \boxed{y=1}$
3. $x + 2y -2z = 4$ $\implies x+2+2=4\implies \boxed{x=0}$

### Exercise 3

Given: $$\begin{cases}
x + y - z - t = 0 \\
2x + 3y - 2z + t = 4 \\
3x + 5z = 0 \\
-x + y - 3z + 2t = 3
\end{cases}$$
Augmented matrix:

$$\begin{bmatrix}
1 & 1 & -1 & -1 & | & 0 \\
2 & 3 & -2 & 1 & | & 4 \\
3 & 0 & 5 & 0 & | & 0 \\
-1 & 1 & -3 & 2 & |& 3
\end{bmatrix}$$

1. $R_{2} - 2R_{1}$: $$\begin{bmatrix}
1 & 1 & -1 & -1 & | & 0 \\
0 & 1 & 0 & 3 & | & 4 \\
3 & 0 & 5 & 0 & | & 0 \\
-1 & 1 & -3 & 2 & |& 3
\end{bmatrix}$$

2. $R_{3} - 3R_{1}$: $$\begin{bmatrix}
1 & 1 & -1 & -1 & | & 0 \\
0 & 1 & 0 & 3 & | & 4 \\
0 & -3 & 8 & 3 & | & 0 \\
-1 & 1 & -3 & 2 & |& 3
\end{bmatrix}$$

3. $R_{4} + R_{1}$: $$\begin{bmatrix}
1 & 1 & -1 & -1 & | & 0 \\
0 & 1 & 0 & 3 & | & 4 \\
0 & -3 & 8 & 3 & | & 0 \\
0 & 2 & -4 & 1 & |& 3
\end{bmatrix}$$

4. $R_{3}+3R_{2}$: $$\begin{bmatrix}
1 & 1 & -1 & -1 & | & 0 \\
0 & 1 & 0 & 3 & | & 4 \\
0 & 0 & 8 & 12 & | & 12 \\
0 & 2 & -4 & 1 & |& 3
\end{bmatrix}$$

5. $R_{4}-2R_{2}$: $$\begin{bmatrix}
1 & 1 & -1 & -1 & | & 0 \\
0 & 1 & 0 & 3 & | & 4 \\
0 & 0 & 8 & 12 & | & 12 \\
0 & 0 & -4 & -5 & |& -5
\end{bmatrix}$$

6. $R_{3}\cdot \frac{1}{8}$: $$\begin{bmatrix}
1 & 1 & -1 & -1 & | & 0 \\
0 & 1 & 0 & 3 & | & 4 \\
0 & 0 & 1 & \frac{3}{2} & | & \frac{3}{2} \\
0 & 0 & -4 & -5 & |& -5
\end{bmatrix}$$

7. $R_{4}+4R_{3}$: $$\begin{bmatrix}
1 & 1 & -1 & -1 & | & 0 \\
0 & 1 & 0 & 3 & | & 4 \\
0 & 0 & 1 & \frac{3}{2} & | & \frac{3}{2} \\
0 & 0 & 0 & 1 & |& 1
\end{bmatrix}$$

Therefore:

1. $\boxed{t = 1}$
2. $z + \frac{3}{2}t = \frac{3}{2}t \implies z+ \frac{3}{2}=\frac{3}{2} \implies$ $\boxed{z=0}$
3. $y + t = 4\implies y + 1 = 4 \implies$ $\boxed{y = 3}$
4. $x+y-z-t=0\implies x+3-0-1=0\implies$ $\boxed{x=-2}$
